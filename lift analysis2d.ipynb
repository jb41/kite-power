```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#data2d = pd.read_csv("debug2d.csv")
data2d = pd.read_csv("debug2d.csv")
hdata2d = pd.read_csv("hdebug2d.csv")

#print(data2d.tail(2))
#print(data3d.tail(2))
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8,6))
ax.plot(data2d["i"], data2d["Lift"], label="2d", linewidth=4)
ax.plot(hdata2d["i"], hdata2d["Lift"], label="h2d", linewidth=2)
ax.legend()

#plt.savefig('et_fisso_neg1_alpha3_ventox_30_mu_0_lift_totale.png')
```
