```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data2d = pd.read_csv("debug2d.csv")
data3d = pd.read_csv("debug3d.csv")
hdata3d = pd.read_csv("hdebug3d.csv")

#print(data2d.tail(2))
#print(data3d.tail(2))
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8,6))
ax.plot(data2d["i"], data2d["Lift"], label="2d", linewidth=4)
ax.plot(data3d["i"], data3d["Lift"], label="3d", linewidth=4)
ax.plot(hdata3d["i"], hdata3d["Lift"], label="h3d", linewidth=2)
ax.legend()

#plt.savefig('esempio discrepanza_alpha12_ventox_30_mu_0_lift_totale.png')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8,6))
ax.plot(data2d["i"], data2d["Liftx"], label="2d", linewidth=4)
ax.plot(hdata3d["i"], hdata3d["Liftx"], label="h3d", linewidth=5)
ax.plot(data3d["i"], data3d["Liftx"], label="3d", linewidth=2)
ax.legend()

#plt.savefig('esempio_oscillazioni_alpha_ventox_30_mu_5_lift_x.png')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8,6))
ax.plot(hdata3d["i"], hdata3d["Lifty"], label="h3d", linewidth=4)
ax.plot(data3d["i"], data3d["Lifty"], label="3d", linewidth=2)
ax.legend()

#plt.savefig('esempio discrepanza_alpha4_ventox_30_mu_5_lifty.png')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8,6))
ax.plot(data2d["i"], data2d["Liftz"], label="2d", linewidth=4)
ax.plot(hdata3d["i"], hdata3d["Liftz"], label="h3d", linewidth=4)
ax.plot(data3d["i"], data3d["Liftz"], label="3d", linewidth=2)
ax.set_xlim(-2,16000)
ax.legend()


#plt.savefig('esempio discrepanza_alpha2_ventox_30_mu0.png')
```
